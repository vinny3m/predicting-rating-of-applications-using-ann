```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
path = r"D:\attachments\playstore.csv"
```

```python
data = pd.read_csv(path)
```

```python
data
```

```python
data.head()
```

```python
data.Size = data.Size.apply(lambda i:i[:-1])
```

```python
data.Installs = data.Installs.apply(lambda i:i[:-1])
```

```python
data.Installs = data.Installs.str.replace(',','')
data.Installs = pd.to_numeric(data.Installs, errors = "coerce")
```

```python
data.Price = data.Price.apply(lambda i:i[1:])
```

```python
data.ContentRating = data.ContentRating.apply(lambda i:i[:-1])
```

```python
k = pd.DataFrame(data)
```

```python
k.head()
```

```python
k.shape
```

```python
k.isnull()
```

```python
m = k['Rating'].mean()
```

```python
k = k.fillna(value = m)
```

```python
k
```

```python
x = k.iloc[:,:6]
```

```python
x.head()
```

```python
y = k.iloc[:,6]
```

```python
y.head()
```

```python
x.describe()
```

```python
x.info()
```

```python
x.shape
```

```python
y.shape
```

```python
# Regression
```

```python
import keras
from keras.models import Sequential
model = Sequential()
```

```python
# Input layer
```

```python
from keras.layers import Dense  
```

```python
model.add(Dense(10,input_shape=(6,)))
```

```python
# Hidden layer
```

```python
model.add(Dense(10))

model.add(Dense(15))
```

```python
# Output layer
```

```python
model.add(Dense(1))
```

```python
# Compile
```

```python
model.compile(loss = "mean_squared_error", optimizer = 'adam') 
```

```python
# Training model
```

```python
model.fit(x, y, epochs = 1000)
```

```python
y_pred = model.predict(x)
```

```python
print(y[5:10])
print(y_pred[5:10])
```

```python
fig, ax = plt.subplots()
ax.scatter(y,y_pred)
ax.plot([y.min(),y.max()], [y.min(),y.max()], 'k--',lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
